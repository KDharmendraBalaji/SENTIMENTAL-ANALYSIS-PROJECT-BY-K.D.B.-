```python
import cv2
from deepface import DeepFace
import os

# Load pre-trained face cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Open a connection to the default camera (0 represents the default camera, change it if you have multiple cameras)
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert the frame to grayscale for face detection
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Analyze emotions for each detected face
    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]

        # Save the face image to a temporary file
        temp_path = "temp.jpg"
        cv2.imwrite(temp_path, face_img)

        try:
            # Analyze emotions for the saved face image with enforce_detection set to False
            emotion_results = DeepFace.analyze(img_path=temp_path, actions=['emotion'], enforce_detection=False)
            
            # If results is a list, take the first element
            if isinstance(emotion_results, list) and emotion_results:
                emotion_result = emotion_results[0]
            else:
                emotion_result = emotion_results

            print(emotion_result)  # Print the result to identify the issue

            # Access the emotion information
            emotion = emotion_result['dominant_emotion']

            # Draw rectangles and display emotion
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
            cv2.putText(frame, f'Emotion: {emotion}', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        except ValueError as e:
            print(f"Error: {e}")

        # Remove the temporary file
        os.remove(temp_path)

    # Display the resulting frame
    cv2.imshow('Emotion Detection', frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cap.release()
cv2.destroyAllWindows()
```
